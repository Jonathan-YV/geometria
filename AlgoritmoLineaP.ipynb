```
import matplotlib.pyplot as plt
```

```
print("Datos de punto P0")
x0=float(input("Escribe la coordenada x0: "))
y0=float(input("Escribe la coordenada y0: "))
print("\nDatos de punto P1")
x1=float(input("Escribe la coordenada x1: "))
y1=float(input("Escribe la coordenada y1: "))
l=float(input("\nSelecciona el valor de lambda (0,1): "))
xT=[]
yT=[]
i=0
x=0
while x<x1:
  x=x0+(i*l)*(x1-x0)
  xT.append(x)
  i=i+1
  if x>x1:
    xT.pop()
i=0
y=0
while y<y1:
  y=y0+(i*l)*(y1-y0)
  yT.append(y)
  i=i+1
  if y>y1:
    yT.pop()
plt.plot(xT,yT,'.')
```

```
import math as m
import matplotlib.pyplot as plt
```

```
x1=int(input("Ingrese la coordenada en x: "))
y1=int(input("Ingrese la coordenada en y: "))
r=int(input("Ingrese el valor del radio: "))
```

```
l=float(input("Ingrese el valor de landa: "))
```

```
angulo=0
xG=[]
yG=[]
while angulo<(2*m.pi):
  xT=x1+r*m.cos(angulo)
  xG.append(xT)
  yT=y1+r*m.sin(angulo)
  yG.append(yT)
  angulo=angulo+l
plt.plot(xG,yG,'.')
plt.axhline(y=0, color='k', linestyle='-')
plt.axvline(x=0,color='k', linestyle='-')
plt.axis('square')
if x1>=0 and y1>=0:
  plt.axis([-2*r, x1+2*r, -2*r, y1+2*r])
if x1>=0 and y1<=0:
  plt.axis([-2*r, x1+2*r, y1-2*r, 2*r])
if x1<=0 and y1>=0:
  plt.axis([x1-2*r, 2*r, -2*r, y1+2*r])
if x1<=0 and y1<=0:
  plt.axis([x1-2*r, 2*r, y1-2*r, 2*r])
```
